```python
page=bs(response.content,'html.parser')
```

```python
target=page.findAll('ul',attrs={'class':'auto_archive'})[0]
```

```python
date=[i.text for i in target.findAll('a')]
```

```python
url=['https://www.who.int'+i.get('href') for i in target.findAll('a')]
```

```python
for i in range(len(url)):
    print(i)
    response=session.get(url[i])
    f=open(f'{date[i]}.html','wb')
    f.write(response.content)
    f.close()
```
